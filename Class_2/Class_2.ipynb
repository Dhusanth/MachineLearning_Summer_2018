#question
define a function so it generates 3 objects with probability 0.3,0.2,0.5 when called

```python
def generate_random():
    r = np.random.random()
    
    if r<0.3:
        return 'A'
    elif r>0.3 and r<0.5:
        return 'B'
    else:
        return 'C'
```

```python
n = 10000

a , b ,c =0,0,0
for i in range(100):
    k = generate_random()
    if k is 'A':
        a +=1
    elif k is 'B':
        b +=1
    else:
        c+= 1
        
print float(a)/n, float(b)/n , float(c)/n
```
